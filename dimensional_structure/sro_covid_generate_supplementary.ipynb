### Supplemental Figures for *Relating psychiatric symptoms and self-regulation during the COVID-19 crisis*

This notebook contains loadings matrices for Exploratory Factor Analysis conducted on training datasets for: 
  * Self-regulation tasks
  * Self-regulation surveys
  * Self-reported psychiatric symptoms
  * Corona Health and Impact Survey
  * Health Risk Behaviors


```python
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from selfregulation.utils.result_utils import load_results
from dimensional_structure.EFA_plots import plot_factor_df
```

```python
all_results  = load_results(datafile = 'Covid_train', verbose = False)
```

### Self-regulation tasks Exploratory Factor Analysis Loadings
4 factors were determined using a BIC criteria for exploratory factor analysis. The training dataset was represented by 386 subjects who completed the battery of tasks on self-regulation only before the onset of the pandemic (i.e., 2016) and part of the dataset published by Eisenberg et al., 2019. The 113 variables are grouped and ordered based on the largest (absolute) factor loading for that variable. **However over the cells to see the exact value**

```python
fig = plt.figure(figsize=(12, 3))
ax1 = fig.add_axes([0.05, 0.80, 0.9, 0.15])
cmap = sns.diverging_palette(220,15,n=20, as_cmap=True)
cb1 = mpl.colorbar.ColorbarBase(ax1, cmap=cmap,
                                orientation='horizontal')
cb1.set_ticks([0,.5,1])
cb1.set_ticklabels(['-1','0', '1'])
cb1.set_label('Color Bar for Factor Loading Plots')
```

```python
plot_factor_df(all_results['task'], 'task')
```

### Self-reported psychiatric symptoms Exploratory Factor Analysis Loadings
To obtain a factorial solution for transdiagnostic psychiatric dimensions, we took advantage of publicly available data (N = 497) collected by independent researchers using the same psychiatric questionnaires we included in our study (Rouault et al., 2018). We extracted 3 factors to obtain transdiagnostic dimensions of psychiatric symptoms as previous studies found that a model with three underlying factors (labelled anxious-depression, compulsive behavior and intrusive thought, and social withdrawal) provided the best account of the covariance for the dataset used here according to the Cattells criterion (Rouault et al., 2018). The 207 variables are grouped and ordered based on the largest (absolute) factor loading for that variable. **However over the cells to see the exact value**




```python
plot_factor_df(all_results['psych'], 'psych')
```

### Corona Health and Impact Survey Exploratory Factor Analysis Loadings
We used a large pool of available data (N = 2868) (Nikolaidis et al., 2021) to derive a factorial solution of the Corona Health and Impact Survey (CRISIS) aimed at investigating several aspects related to the emergence of the pandemic such as worries related to COVID-19 or the subjective impact of life changes associated to it. 
10 factors were determined using a BIC criteria for exploratory factor analysis. The 37 variables are grouped and ordered based on the largest (absolute) factor loading for that variable. **However over the cells to see the exact value**

```python
plot_factor_df(all_results['crisis'], 'crisis')
```
